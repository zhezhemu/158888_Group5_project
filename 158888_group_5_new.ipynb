```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
df = pd.read_csv('ads_trend_analysis.csv')
df.sort_values(by=['Country', 'year'], inplace=True)
all_predictions = []
for country_code in df['Country_Code_ISO3'].unique():
    country_data = df[df['Country_Code_ISO3'] == country_code]
    if country_data.empty:
        print(f"No data for country code: {country_code}")
        continue
    for variable in ['Goal_8_Score', 'sdg8_unemp', 'sdg8_adjgrowth']:
        ts = country_data[variable].dropna()
        if ts.empty or len(ts) < 5:
            print(f"Not enough data for {variable} in {country_code}")
            continue
        model = ARIMA(ts, order=(1,1,1))
        model_fit = model.fit()
        forecast = model_fit.forecast(steps=5)
        forecast_years = list(range(country_data['year'].max() + 1, country_data['year'].max() + 6))
        predictions = {
            'Country_Code_ISO3': country_code,
            'Country': country_data['Country'].iloc[0],
            'year': forecast_years,
            'variable': [variable] * 5,
            'predicted_value': forecast
        }
        all_predictions.append(pd.DataFrame(predictions))
final_predictions_df = pd.concat(all_predictions)
final_predictions_df = final_predictions_df.pivot_table(index=['Country_Code_ISO3', 'Country', 'year'],columns='variable',values='predicted_value').reset_index()
final_predictions_df.to_csv('predictions.csv', index=False)
print("The predictions have been saved to predictions.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'SDR_2023_Data.xlsx'
data = pd.read_excel(file_path)
numeric_data = data.select_dtypes(include=[np.number]).fillna(data.mean(numeric_only=True))
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)
pca = PCA(n_components=0.9)
pca_data = pca.fit_transform(scaled_data)

plt.figure(figsize=(10, 8))
sns.scatterplot(x=pca_data[:, 0], y=pca_data[:, 1])
plt.title('Scatter Plot of Data by First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)
plt.grid(True)
plt.show()

wcss = []
silhouette_coefficients = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(pca_data)
    wcss.append(kmeans.inertia_)
    silhouette_coefficients.append(silhouette_score(pca_data, kmeans.labels_))

plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.plot(range(2, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')

# Plotting the Silhouette Coefficients
plt.subplot(1, 2, 2)
plt.plot(range(2, 11), silhouette_coefficients, marker='o')
plt.title('Silhouette Coefficients')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Coefficient')

plt.tight_layout()
plt.show()

optimal_clusters = 4
kmeans_optimal = KMeans(n_clusters=optimal_clusters, init='k-means++', random_state=42)
cluster_labels = kmeans_optimal.fit_predict(pca_data)

plt.figure(figsize=(10, 8))
sns.scatterplot(x=pca_data[:, 0], y=pca_data[:, 1], hue=cluster_labels, palette='viridis', legend='full')
plt.title('Cluster Visualization by First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)
plt.grid(True)
plt.legend(title='Cluster')
plt.show()
```

```python
import pandas as pd
import numpy as np

df = pd.read_excel('SDR_2023_Data.xlsx')
indicators = [
    "Normalized Score: sdg8_adjgrowth",
    "Normalized Score: sdg9_intuse",
    "Normalized Score: sdg9_mobuse",
    "Normalized Score: sdg9_lpi",
    "Normalized Score: sdg9_rdex"
]
df_selected = df[["Country Code ISO3", "Country"] + indicators]
df_selected[indicators] = df_selected[indicators].apply(pd.to_numeric, errors='coerce')
df_selected = df_selected.dropna(subset=indicators)
df_std = df_selected.copy()
df_std[indicators] = (df_std[indicators] - df_std[indicators].min()) / (df_std[indicators].max() - df_std[indicators].min())
pj = df_std[indicators].div(df_std[indicators].sum(axis=0), axis=1)
k = 1.0 / np.log(len(df_std))
entropy = -k * (pj * np.log(pj + 1e-10)).sum(axis=0)
diversity = 1 - entropy
weights = diversity / diversity.sum()
sij = df_std[indicators].multiply(weights, axis=1)
df_selected['Economic Growth Index'] = sij.sum(axis=1)
economic_growth_index = df_selected[["Country Code ISO3", "Country", "Economic Growth Index"]]
weights_df = pd.DataFrame(weights, columns=['Weight']).reset_index()
weights_df, economic_growth_index
```
